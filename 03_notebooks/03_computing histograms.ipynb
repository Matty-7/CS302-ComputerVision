```python
import numpy as np

from matplotlib import pyplot as plt
plt.rcParams['axes.labelsize'] = 20
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
%matplotlib inline 
```

```python
from skimage.io import imread
astro_bw = imread('astronaut_bw.png')
```

## computing histograms

```python
plt.figure(figsize=(19,12))

plt.subplot(221)
plt.imshow(astro_bw, cmap = plt.cm.gray)
plt.axis('off')
plt.title('Collins', fontsize=20)

plt.subplot(222)
plt.hist(astro_bw.flatten(), bins=256, range=(0,256))  # range is important if min/max are not 0/255
plt.ylabel('number of pixels')
plt.xlabel('gray value')
plt.title('histogram', fontsize=20)

plt.subplot(223)
plt.hist(astro_bw.flatten(), bins=256, range=(0,256) ,align = 'left')  # align for centering
plt.ylabel('number of pixels')
plt.xlabel('gray value')
plt.xlim(245, 256)
plt.ylim(0,1600)
plt.title('histogram, zoom', fontsize=20)

plt.subplot(224)
plt.hist(astro_bw.flatten(), bins=256, range=(0,256), cumulative=True, density=True)
plt.xlabel('gray value')
plt.title('cummulative histogram', fontsize=20)

plt.show()
```
