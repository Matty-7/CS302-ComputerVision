```python
import numpy as np

from matplotlib import pyplot as plt
plt.rcParams['axes.labelsize'] = 20
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
%matplotlib inline 
```

```python
from skimage.io import imread

wedge = imread('al_wedge.png')   #image source: Paulo Vinícius Soares

height, width = wedge.shape
print('the image has {} rows and {} columns'.format(height,width))

plt.figure(figsize=(14,7))
plt.imshow(wedge, cmap=plt.cm.gray)
# plt.axis('off')   # <- we want to see the pixel coordinates
plt.show()
```

## Read a line

```python
row = 170

# read out the gray values
gv_row = wedge[row,:]

# paint black lines into a new image
visual= np.copy(wedge)   # allocate new memory
visual[row,:] = 0

# get array of x-values for print
x_row = np.arange(width)

plt.figure(figsize=(21,8))

plt.subplot(121)
plt.imshow(visual, cmap=plt.cm.gray)

plt.subplot(122)
plt.plot(x_row, gv_row, 'b-')
plt.xlabel('pixel')
plt.ylabel('gray values')

plt.show()
```

If you can not see the black line, it probably became invisible while the image was scaled. Try changing the row number or the figure size.
