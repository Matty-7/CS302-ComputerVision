```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from scipy import datasets

# load our default image
face = datasets.face(gray=True)
```

## Averaging

```python
from scipy import ndimage

# apply averaging filter
average_face = ndimage.uniform_filter(face, size=11)
```

```python
# plot results
plt.figure(figsize=(17, 7))
plt.subplot(121)
plt.imshow(face, cmap=plt.cm.gray)
plt.axis('off')
plt.subplot(122)
plt.imshow(average_face, cmap=plt.cm.gray)
plt.axis('off')

plt.show()
```

## Gaussian filter

```python
from skimage import filters
Gaussian_face  = filters.gaussian(face, sigma=3)
```

```python
plt.figure(figsize=(17, 7))

plt.subplot(121)
plt.imshow(face, cmap=plt.cm.gray)
plt.axis('off')
plt.title('raw', fontsize=20)

plt.subplot(122)
plt.imshow(Gaussian_face, cmap=plt.cm.gray)
plt.axis('off')
plt.title('Gaussian', fontsize=20)

plt.show()
```

## visualizing the difference

```python
face_small = face[300:400, 300:500]
Gaussian_face_small  = filters.gaussian(face_small, sigma=3)
averaged_face_small = ndimage.uniform_filter(face_small, size=11)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(131)
plt.imshow(face_small, cmap=plt.cm.gray)
plt.axis('off')
plt.title('raw', fontsize=20)

plt.subplot(132)
plt.imshow(Gaussian_face_small, cmap=plt.cm.gray)
plt.axis('off')
plt.title('Gaussian', fontsize=20)

plt.subplot(133)
plt.imshow(averaged_face_small, cmap=plt.cm.gray)
plt.axis('off')
plt.title('averaged', fontsize=20)

plt.show()
```
