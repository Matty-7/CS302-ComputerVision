```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
%matplotlib inline         

from skimage.io import imread
from skimage import color
```

```python
astro = imread('astronaut.png')

plt.figure(figsize=(6,6))
plt.imshow(astro)
plt.axis('off')
plt.title('Eileen Collins', fontsize=20)
plt.show()
```

```python
astro.shape
```

```python
# separate image into its three channels
red = astro[:,:,0]
green = astro[:,:,1]
blue = astro[:,:,2]
```

```python
plt.figure(figsize=(18,8))

plt.subplot(131)
plt.imshow(red, cmap=plt.cm.gray)
plt.axis('off')
plt.title('red', fontsize=20)

plt.subplot(132)
plt.imshow(green, cmap=plt.cm.gray)
plt.axis('off')
plt.title('green', fontsize=20)

plt.subplot(133)
plt.imshow(blue, cmap=plt.cm.gray)
plt.axis('off')
plt.title('blue', fontsize=20)

plt.show()
```

```python
#invert the green channel
green_inv = 255-green

# concatenate three channels to an RGB image
RgB = np.dstack((red, green_inv, blue))
```

```python
plt.figure(figsize=(8,8))

plt.imshow(RgB)
plt.axis('off')
plt.title('green inverted', fontsize=20)

plt.show()
```
