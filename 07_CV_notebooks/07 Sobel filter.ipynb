
Finding edges with Sobel filters
=================================


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from scipy import datasets
from skimage import filters
```

```python
# we need to convert the image to floating point
face_int = datasets.face(gray=True)

from skimage import img_as_float
face = img_as_float(face_int)
```

```python
# apply Sobel filter 
sx = filters.sobel(face, axis=0)
sy = filters.sobel(face, axis=1) 

sob = np.hypot(sx, sy)  #hypotenuse of the triangle
```

```python
# range of values
print('sx range: {:.3} to {:.3}'.format(sx.min(), sx.max()))
print('sob range: {:.3} to {:.3}'.format(sob.min(), sob.max()))
```

```python
# plot result
plt.figure(figsize=(17, 14))
plt.subplot(221)
plt.imshow(face, cmap=plt.cm.gray)
plt.axis('off')
plt.title('raw', fontsize=20)

plt.subplot(222)
plt.imshow(sx, cmap=plt.cm.gray)
plt.axis('off')
plt.title('Sobel, horizontal', fontsize=20)

plt.subplot(223)
plt.imshow(sy, cmap=plt.cm.gray)
plt.axis('off')
plt.title('Sobel vertical', fontsize=20)

plt.subplot(224)
plt.imshow(sob, cmap=plt.cm.gray)
plt.axis('off')
plt.title('Sobel filter', fontsize=20)

plt.show()
```
